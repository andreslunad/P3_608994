```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

def save_mnist_to_dirs(images, labels, base_dir, prefix='mnist'):
    for i, (img, lab) in enumerate(zip(images, labels)):
        cls_dir = os.path.join(base_dir, str(int(lab)))
        os.makedirs(cls_dir, exist_ok=True)
        fname = os.path.join(cls_dir, f"{prefix}_{i:05d}.png")
        cv2.imwrite(fname, img)

# Guardar MNIST
save_mnist_to_dirs(train_images, train_labels, os.path.join(mnist_root, 'train'), prefix='mnist_tr')
save_mnist_to_dirs(test_images, test_labels, os.path.join(mnist_root, 'test'), prefix='mnist_ts')
```

```python
BATCH = 32

# =====================================
# PARÁMETROS PARA ENTRENAMIENTO RÁPIDO
# =====================================
EPOCHS_EXP = 4          # épocas para modelos exploratorios
PATIENCE_ES = 2         # paciencia para early stopping
FAST_FRACTION = 0.25    # fracción de dataset por epoch

train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1
)

train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=IMG_SIZE,
    color_mode='rgb',
    class_mode='categorical',
    batch_size=BATCH,
    shuffle=True,
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=IMG_SIZE,
    color_mode='rgb',
    class_mode='categorical',
    batch_size=BATCH,
    shuffle=False,
    subset='validation'
)
```

```python
# ==========================================================
# 5 MODELOS ADICIONALES PARA EL PUNTO 3 DEL PROYECTO
# ==========================================================

from tensorflow.keras import layers, models

# Generadores rápidos grayscale 28x28
IMG_GRAY = (28,28)

gray_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_gray = gray_datagen.flow_from_directory(
    train_path, target_size=IMG_GRAY, color_mode='grayscale',
    class_mode='categorical', batch_size=32, subset='training'
)

val_gray = gray_datagen.flow_from_directory(
    train_path, target_size=IMG_GRAY, color_mode='grayscale',
    class_mode='categorical', batch_size=32, subset='validation'
)

# ----------------------------
# Modelo 1 – CNN muy pequeño
# ----------------------------
def get_model1():
    m = models.Sequential([
        layers.Input(shape=(28,28,1)),
        layers.Conv2D(16,3,activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(32,activation='relu'),
        layers.Dense(10,activation='softmax')
    ])
    m.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
    return m

# ----------------------------
# Modelo 2 – CNN más profundo
# ----------------------------
def get_model2():
    m = models.Sequential([
        layers.Input(shape=(28,28,1)),
        layers.Conv2D(32,3,activation='relu'),
        layers.MaxPooling2D(),
        layers.Conv2D(64,3,activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(64,activation='relu'),
        layers.Dense(10,activation='softmax')
    ])
    m.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
    return m

# ----------------------------
# Modelo 3 – CNN con BatchNorm
# ----------------------------
def get_model3():
    m = models.Sequential([
        layers.Input(shape=(28,28,1)),
        layers.Conv2D(32,3,padding="same"), layers.BatchNormalization(), layers.Activation("relu"),
        layers.MaxPooling2D(),
        layers.Conv2D(64,3,padding="same"), layers.BatchNormalization(), layers.Activation("relu"),
        layers.GlobalAveragePooling2D(),
        layers.Dense(64,activation='relu'),
        layers.Dense(10,activation='softmax')
    ])
    m.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
    return m

# ----------------------------
# Modelo 4 – CNN con Dropout
# ----------------------------
def get_model4():
    m = models.Sequential([
        layers.Input(shape=(28,28,1)),
        layers.Conv2D(32,3,activation='relu'),
        layers.MaxPooling2D(),
        layers.Dropout(0.2),
        layers.Conv2D(64,3,activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(128,activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(10,activation='softmax')
    ])
    m.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
    return m

# ----------------------------
# Modelo 5 – CNN con data augmentation fuerte
# ----------------------------
aug_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=20,
    width_shift_range=0.15,
    height_shift_range=0.15,
    zoom_range=0.2
)

train_aug = aug_datagen.flow_from_directory(
    train_path, target_size=IMG_GRAY, color_mode='grayscale',
    class_mode='categorical', batch_size=32, subset='training'
)

val_aug = aug_datagen.flow_from_directory(
    train_path, target_size=IMG_GRAY, color_mode='grayscale',
    class_mode='categorical', batch_size=32, subset='validation'
)

def get_model5():
    m = models.Sequential([
        layers.Input(shape=(28,28,1)),
        layers.Conv2D(32,3,activation='relu'),
        layers.MaxPooling2D(),
        layers.Conv2D(64,3,activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(64,activation='relu'),
        layers.Dense(10,activation='softmax')
    ])
    m.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
    return m


# ---- ENTRENAMIENTO DE LOS 5 MODELOS ----

histories = {}

models_list = [
    ("Modelo_1", get_model1(), train_gray, val_gray),
    ("Modelo_2", get_model2(), train_gray, val_gray),
    ("Modelo_3_BN", get_model3(), train_gray, val_gray),
    ("Modelo_4_Dropout", get_model4(), train_gray, val_gray),
    ("Modelo_5_Augment", get_model5(), train_aug, val_aug),
]

# ==========================================================
# ENTRENAMIENTO RÁPIDO DE LOS 5 MODELOS
# ==========================================================
from tensorflow.keras.callbacks import EarlyStopping

def reduced_steps(gen, frac=FAST_FRACTION):
    return max(1, int((gen.n // gen.batch_size) * frac))

es_fast = EarlyStopping(
    monitor='val_accuracy',
    patience=PATIENCE_ES,
    restore_best_weights=True
)

for name, mdl, tr, val in models_list:
    print("\n🚀 Entrenando (rápido):", name)

    steps = reduced_steps(tr)
    val_steps = reduced_steps(val)

    h = mdl.fit(
        tr,
        validation_data=val,
        epochs=EPOCHS_EXP,
        steps_per_epoch=steps,
        validation_steps=val_steps,
        callbacks=[es_fast],
        verbose=1
    )
    histories[name] = h
```

```python
from tensorflow.keras.applications import MobileNetV2

base_model = MobileNetV2(input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3),
                         include_top=False,
                         weights='imagenet')
base_model.trainable = False

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
EPOCHS = 20

early_stop = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.h5', monitor='val_accuracy', save_best_only=True)

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)
```

```python
# Descongelar últimas 20 capas
base_model.trainable = True
for layer in base_model.layers[:-20]:
    layer.trainable = False

# Recompilar con LR menor
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history_fine = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,
    callbacks=[early_stop, checkpoint]
)
```

```python
# === SOLUCIÓN PARA KERAS 3 COMPATIBLE CON WINDOWS Y TFLITE ===
print("Guardando tu modelo entrenado de forma compatible...")

import tensorflow as tf

# USAR EL MODELO GANADOR ENTRENADO CON TODO EL CONJUNTO
winner = model_full


# --- 1) Convertir profundamente todo el modelo a float32 ---
def convert_model_to_float32(model):
    # Cambiar dtype de cada capa
    for layer in model.layers:
        try:
            layer.dtype = "float32"
        except:
            pass
        try:
            layer._dtype = tf.float32
        except:
            pass

    # Reconstruir modelo con entrada float32
    new_model = tf.keras.models.clone_model(
        model,
        input_tensors=tf.keras.Input(shape=model.input_shape[1:], dtype="float32")
    )
    new_model.set_weights(model.get_weights())
    return new_model


winner_float32 = convert_model_to_float32(winner)
print("✔ Modelo convertido internamente a float32")


# --- 2) Guardar modelo H5 (compatible Windows) ---
winner_float32.save("modelo_compatible.h5", include_optimizer=False)
print("✔ Modelo guardado como .h5 (float32 compatible)")


converter = tf.lite.TFLiteConverter.from_keras_model(winner)
converter.target_spec.supported_ops = [
    tf.lite.OpsSet.TFLITE_BUILTINS,
    tf.lite.OpsSet.SELECT_TF_OPS
]
tflite_model = converter.convert()
open("modelo_compatible.tflite", "wb").write(tflite_model)



# --- 4) Guardar formato .keras opcional ---
winner_float32.save('modelo_compatible.keras')
print("✔ Modelo guardado como .keras (float32)")


# --- 5) Guardar arquitectura + pesos como respaldo ---
winner_json = winner_float32.to_json()
with open('model_arquitectura.json', 'w') as f:
    f.write(winner_json)

winner_float32.save_weights('model_pesos.weights.h5')
print("✔ Arquitectura y pesos guardados por separado")

print("\n🎯 Listo: todos los formatos fueron creados 100% compatibles.")
```

```python
%%writefile realtime_multiformat.py
import cv2
import numpy as np
import tensorflow as tf

def cargar_modelo_multiple():
    """Intenta cargar el modelo en diferentes formatos"""

    # Intentar formato .keras (Keras 3)
    try:
        model = tf.keras.models.load_model("modelo_compatible.keras", compile=False)
        print("✔ Modelo cargado desde formato .keras")
        return model
    except:
        pass

    # Intentar formato .h5
    try:
        model = tf.keras.models.load_model("modelo_compatible.h5", compile=False)
        print("✔ Modelo cargado desde formato .h5")
        return model
    except:
        pass

    # Intentar desde arquitectura + pesos
    try:
        with open("model_arquitectura.json", "r") as f:
            model_json = f.read()
        model = tf.keras.models.model_from_json(model_json)
        model.load_weights("model_pesos.weights.h5")
        print("✔ Modelo cargado desde arquitectura + pesos")
        return model
    except Exception as e:
        print(f"❌ Error cargando modelo: {e}")
        return None

def cargar_modelo_tflite():
    """Carga el modelo en formato TensorFlow Lite"""
    try:
        interpreter = tf.lite.Interpreter(model_path="modelo_compatible.tflite")
        interpreter.allocate_tensors()
        input_details = interpreter.get_input_details()
        output_details = interpreter.get_output_details()
        print("✔ Modelo TensorFlow Lite cargado")
        return interpreter, input_details, output_details
    except Exception as e:
        print(f"❌ Error cargando TFLite: {e}")
        return None, None, None

# PRIMERO intentar con modelo normal
print("Intentando cargar modelo Keras...")
model = cargar_modelo_multiple()

if model is not None:
    use_tflite = False
    print("✅ Usando modelo Keras normal")
else:
    # Si falla, usar TFLite
    print("Intentando cargar modelo TensorFlow Lite...")
    interpreter, input_details, output_details = cargar_modelo_tflite()
    if interpreter is not None:
        use_tflite = True
        print("✅ Usando modelo TensorFlow Lite")
    else:
        print("❌ No se pudo cargar ningún modelo")
        exit()

IMG_SIZE = (96, 96)

# Inicializar cámara
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("❌ No se pudo acceder a la cámara")
    exit()

print("🎥 Cámara iniciada. Presiona 'q' para salir")

while True:
    ret, frame = cap.read()
    if not ret:
        print("❌ Error leyendo frame de cámara")
        break

    # Preprocesamiento IDÉNTICO a tu entrenamiento
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    inp = cv2.resize(gray, IMG_SIZE)
    inp_norm = inp.astype('float32') / 255.0

    # Convertir a RGB (igual que en tu entrenamiento)
    inp_rgb = np.stack([inp_norm, inp_norm, inp_norm], axis=-1)
    x = np.expand_dims(inp_rgb, axis=0)

    try:
        if not use_tflite:
            # Predecir con modelo Keras normal
            probs = model.predict(x, verbose=0)[0]
        else:
            # Predecir con TFLite
            interpreter.set_tensor(input_details[0]['index'], x)
            interpreter.invoke()
            probs = interpreter.get_tensor(output_details[0]['index'])[0]

        winner = np.argmax(probs)
        confidence = probs[winner]

        # Mostrar probabilidades
        for i, p in enumerate(probs):
            color = (0, 255, 0) if i == winner else (0, 0, 255)
            text = f"{i}: {p:.3f}"
            cv2.putText(frame, text, (10, 30 + i * 25),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

        # Mostrar predicción
        cv2.putText(frame, f"Pred: {winner} ({confidence:.2f})",
                   (10, 300), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)

        # Mostrar input del modelo
        model_input_display = cv2.resize(inp_rgb, (100, 100)) * 255
        model_input_display = model_input_display.astype('uint8')
        frame[10:110, 400:500] = model_input_display
        cv2.putText(frame, "Input al modelo", (400, 120),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

    except Exception as e:
        cv2.putText(frame, f"Error: {str(e)}", (10, 30),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)

    cv2.imshow("Reconocimiento de Digitos - Presiona 'q' para salir", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
print("✅ Programa terminado")
```
